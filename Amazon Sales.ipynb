```python
df['Shipping_Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df['Profit_Margin_%'] = (df['Total Profit'] / df['Total Revenue']) * 100
```

```python
df['shipping_category'] = pd.cut(df['Shipping_Days'], bins=[-1, 5, 10, 20, float('inf')], labels=['Super Fast','Fast','Slow','Delayed'])
```

```python
df.columns
```

```python
df['Profit Categoty'] = pd.qcut(df['Total Profit'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
# Are there any negative shipping days?
df[df['Shipping_Days'] < 0]
```

```python
from sqlalchemy import create_engine

server = r'localhost\SQLEXPRESS'
database = 'Amazon_Sales'   

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)

engine = create_engine(connection_string)

print("Connected successfully")
```

```python
# Upload dataframe as SQL table
df.to_sql(
    name='amazon_sales',
    con=engine,
    if_exists='replace',
    index=False
)

print("Amazon sales data uploaded to SQL Server successfully.")
```
